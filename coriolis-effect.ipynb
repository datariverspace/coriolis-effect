```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_circle():
    theta = np.linspace(0, 2 * np.pi, 100)
    
    radius = 0.2
    center_x = 0
    center_y = 0
    
    x = center_x + radius * np.cos(theta)
    y = center_y + radius * np.sin(theta)
   
    plt.figure()
    plt.plot(x, y, color='green')
    plt.axis('equal') # Ensure the circle isn't distorted
    plt.title('The Earth')
    plt.show()

plot_circle()
```
